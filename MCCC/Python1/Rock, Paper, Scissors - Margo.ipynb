# A game of rock, paper, scissors against computer. 

## Set a play choice (rock, paper, or scissors) for the computer player and during the program, request for the user to input their play choice.

#### The rules in Rock, Paper, Scissors are:
    Rock beats scissors. 
    Scissors beats paper. 
    Paper beats rock

#### If the user wins, then display "You win!" If they lose then display "You lose." If the computer and the user have the same play choice, display "It's a draw."

###### BONUS: Randomize the play choice for the computer (hint: Python has a library called random). Also ask the user if they would like to play again. If they do, start the game again but if not, thank the user for playing the game.

```python
import random as rand
```

```python
def choiceMenu():
    print("""Welcome to Rock, Paper, Scissor Game!
    
    The rules:
        Rock beats scissors.
        Scissors beats paper.
        Paper beats rock.
    
    How to Play:
        Enter 'r' to throw 'Rock'.
        Enter 'p' to throw 'Paper'.
        Enter 's' to throw 'Scissor'.
        Enter 'q' to quit game.
    """)

def how(pick):
    if pick == "r":
        print("Rock")
    if pick == "p":
        print("Paper")
    if pick == "s":
        print("Scissor")
        
def play():
    you = input("\nEnter 'r', 'p', 's', or 'q': ")
    you = you.lower
    if you == 'q':
        break
        print(f"""What a great game! Thanks for playing!
        Score: 
    Wins:\t{win}
    Lose:\t{lose}
    Draw:\t{draw}
    """)
    how(you)
    comp = rand.randint(1,3)
    if comp == 1: 
        pick = "r"
    elif comp == 2:
        pick = "p"
    else: 
        pick = "s"
    print(pick)
    how(pick)
    rules(you, comp)
    again = input("Do you want to play again: (y/n): ")
    if again == "n":
        print(f"""What a great game! Thanks for playing!
        Score: 
    Wins:\t{win}
    Lose:\t{lose}
    Draw:\t{draw}
    """)
    else:
        play()
            

def rules(you, comp):
    win = 0
    lose = 0
    draw = 0
    if you == 'p' and comp == 'r':
        win += 1 
        print("You win!")
    elif you == 'r' and comp == 's':
        win += 1
        print("You win!")
    elif you == 's' and comp == 'p':
        win += 1
        print("You win!")
    if you == 'p' and comp == 's':
        lose += 1 
        print("You lose!")
    elif you == 'r' and comp == 'p':
        lose += 1
        print("You lose!")
    elif you == 's' and comp == 'r':
        lose += 1
        print("You lose!")
    else: 
        draw += 1
        print("It's a draw!")
    
while True:
    choiceMenu()
    play()

        
    
```
