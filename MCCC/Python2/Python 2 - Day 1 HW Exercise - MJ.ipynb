## Exercise 1: 
Create a Player class with a name, a position, and a speed attributes. Set the initial value of speed to 1 and create any necessary methods.
Then, 
1. Create two Player objects and print information about both players.
2. Assign 8 to player 1's speed
3. print information about both players.

```python
class Player: 
    """Player has name, position, and speed attributes. Speed has default value '1'"""
    
    def __init__(self, name):
        self.name = name
        self.position = None
        self.speed = 1
       
    def getPostion(self):
        return self.position
    
    def getSpeed(self):
        return self.speed
        
    def info(self): 
        print(f"{self.name} is in position {self.position} going {self.speed} mph.")
    
    def setPosition(self,position):
        self.position += position
    
    def setSpeed(self,speed):
        self.speed += speed
```

```python
p1 = Player("PlayerOne")
p2 = Player("PlayerTwo")
p1.info()
p2.info()
print("<>"*25)
p1.setSpeed(7)
p1.info()
p2.info()
```

## Exercise 3: 
Create a class called Triangle with height and base attributes and has following methods:

- perimeter() : returns a perimeter of a right triangle object
- area() : returns an area of a right triangle object

Then,

- create two instances of Triangle class
- print each triangle's perimeter and determine which triangle has larger perimeter
- print each triangle's area and determine which triangle has larger area

Use 
- perimeter of a right triangle = height + base + Square-Root((height \** 2) + (base \** 2)) 
- area of a right triangle = ( height * base ) / 2

You'll need to <b>import math</b> library and use <b>math.sqrt()</b> function.



```python
import math

class Triangle:
    """Triangle has height and base attributes."""
    
    def __init__(self, height, base):
        self.height = height
        self.base = base
    
    def area(self):
        return (self.height * self.base) / 2

    def areaInfo(self):
        print(f"The area is {((self.height * self.base) / 2):,.2f}.")

    def periInfo(self):
        print(f"The perimeter is {(self.height + self.base + (math.sqrt(self.height) + math.sqrt(self.base))):,.2f}.")

    def perimeter(self):
        return self.height + self.base + (math.sqrt(self.height) + math.sqrt(self.base))
    
Tri1 = Triangle(rand.randint(1,20), rand.randint(1,20))
Tri2 = Triangle(rand.randint(1,20), rand.randint(1,20))

print("Tri1:")
Tri1.perimeter()
Tri1.periInfo()
print()
print("Tri2:")
Tri2.perimeter()
Tri2.periInfo()
print()

if Tri1.perimeter() > Tri2.perimeter():
    print("Tri1 has a larger perimeter.")
else: 
    print("Tri2 has a larger perimeter.")
print("<>"*25)
print()

print("Tri1:")
Tri1.area()
Tri1.areaInfo()
print()
print("Tri2:")
Tri2.area()
Tri2.areaInfo()
print()

if Tri1.area() > Tri2.area():
    print("Tri1 has a larger area.")
else: 
    print("Tri2 has a larger area.")
print("<>"*25)

```
